```python
def remove_rare_words(text):
    a= text.split()
    df = pd.DataFrame({'col': a})
    m = df.groupby('col').filter(lambda x : len(x)>1)
    mo = m['col'].tolist()
    text=''
    for word in mo:
        text=text+word+' '
    return text

```
