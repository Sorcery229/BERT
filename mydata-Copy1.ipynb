```python
# mydata


import tqdm
import requests
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
import re
from bs4 import BeautifulSoup
import cchardet
import multiprocessing
from concurrent.futures import ThreadPoolExecutor

# html = """<p>exact text</p>
#    <p>almost exact text</p>
#    <span class="tm-article-snippet__hubs-item"><a href="/ru/hub/ui/" class="tm-article-snippet__hubs-item-link"><span>Интерфейсы</span> <span title="Профильный хаб" class="tm-article-snippet__profiled-hub">*</span></a></span>
#    <span>Интерфейсы</span>"""
# soup = BeautifulSoup(html)
# print(soup(text='exact text'))
# print(soup(text=re.compile('span')))
# elements=soup.findAll("span", attrs = {'class' : 'tm-article-snippet__hubs-item'})
# for element in elements:
#     print(element.get_text())
habrParse_df = pd.DataFrame(columns=[
                                     'link',
                                     'response_title',
                                     'response_post',
                                     'tag'])
mainUrl = 'https://habr.com/ru/post/'
postCount = 1995
# @ratelim.patient(1, 1)
def get_post(link):
    
    try:
        response = requests.get(link)
#         response.raise_for_status()
        
        soup = bs(response.text, 'lxml')
        # Получаем заголовок статьи
        title = soup.find('meta', property='og:title')
        title = str(title).split('="')[1].split('" ')[0]
        response_title = title
        # Получаем текст статьи
        post = str(soup.find('div', id="post-content-body"))
        post = re.sub('\n', ' ', post)
        response_post=post
        # Получаем тег
    #     tag = str(soup.find(re.compile("<span>*<span>")))

    #     tag = str(soup.find('meta', property='og:title')).split('="')[1].split('" ')[0]

        elements=soup.findAll("span", attrs = {'class' : 'tm-article-snippet__hubs-item'})
        tag=""
        for element in elements:
            tag = tag+element.get_text()+" "
        response_tag = tag
        
        dataList = [link, response_title, response_post, tag]
        habrParse_df.loc[len(habrParse_df)] = dataList
        print(len(habrParse_df))
    except requests.exceptions.HTTPError as err:
        pass
    except requests.exceptions.ConnectionError as err2:
        pass
      

# for pc in tqdm.tqdm(range(postCount)):
#     postNum = pc + 1
#     get_post(postNum,session_object)
    
links=[]
for pc in range(postCount):
    postNum = pc + 1
    links.append(mainUrl+str(234000+postNum))

with ThreadPoolExecutor(max_workers=4) as executor:
    executor.map(get_post,links)

habrParse_df.head()
habrParse_df.to_csv('mydata103.csv')

# habrParse_df.head()
# habrParse_df.to_csv('mydata.csv')
# habrParse_df['response_post'][2]
# habrParse_df.to_csv('mydata2.csv')
# df = pd.read_csv('mydata.csv')
# df['tag'].value_counts().head(40)
# mydata.csv  53+strnum
# mydata1.csv 10000 str(strnum+20000)
# habrParse_df['tag'].value_counts().head(40)
#https://www.paypal.com/smarthelp/contact-us?email=privacy
```
